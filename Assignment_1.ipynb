```python
import pandas as pd
```

```python
df=pd.read_csv('jsrt_metadata.csv')
```

```python
import numpy as np
```

```python
n=df['size'].astype('float').mean()
```

```python
n
```

```python
df['size'].replace(np.nan,n,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```
